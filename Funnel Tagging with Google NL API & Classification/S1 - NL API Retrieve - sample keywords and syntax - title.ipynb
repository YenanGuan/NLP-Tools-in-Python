```python
import pandas as pd
import numpy as np
import time

import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('KWSerpAnalysis 05 - Authority Labs Scrape 2019-09-12.csv')
kw = pd.read_csv('mkt funnel keywords for model.csv')
print(list(kw))

df = df[df['query'].isin(kw['query'])].reset_index(drop=True)

df = df[['query', 'title']]

df.sample(3)
```

## Create API Connection for Syntax Retrieval
- https://cloud.google.com/natural-language/docs/basics#syntactic_analysis
- https://cloud.google.com/natural-language/docs/analyzing-syntax
- https://cloud.google.com/natural-language/docs/reference/rest/v1/Token

```python
from google.cloud import language_v1
from google.cloud.language_v1 import enums
from google.oauth2 import service_account
from googleapiclient.errors import HttpError

credentials = service_account.Credentials.from_service_account_file(
    'credential path.json')

client = language_v1.LanguageServiceClient(credentials=credentials)


def analyze_syntax(text_content):

    document = {"content": text_content,
                "type": enums.Document.Type.PLAIN_TEXT,
                "language": "en"}

    response = client.analyze_syntax(document=document,
                                     encoding_type=enums.EncodingType.UTF8)

    pos = list()
    lemma = list()
    label = list()
    
    for token in response.tokens:

        text = token.text
        part_of_speech = token.part_of_speech
        
        pos.append(enums.PartOfSpeech.Tag(part_of_speech.tag).name)
        
        lemma.append(token.lemma)

        dependency_edge = token.dependency_edge
        label.append(enums.DependencyEdge.Label(dependency_edge.label).name)
    
    return pos, lemma, label
```

## Execute API and Store Results

```python
df['pos'] = None
df['lemma'] = None
df['label'] = None

for i in range(len(df)):

    text = df.iloc[i]['title']
    if type(text) == str:
        run = True
        while(run):
            try:
                pos, lemma, label = analyze_syntax(text)
                run = False
                print('Request for 【', i, '/', len(df), '】', 'completed')
            except:
                print('Limit Error, Sleeping --- Zzz')
                time.sleep(60)
                run = True
                print('Trying again')


        df.loc[i, 'pos'] = pos
        df.loc[i, 'lemma'] = lemma
        df.loc[i, 'label'] = label
    else:
        print('【', i, '/', len(df), '】', 'is nan')
```
